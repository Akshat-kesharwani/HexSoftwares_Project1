```python
2 #Relationship Data Analysis & Visualization:-
```

```python
#Heatmap:-
```

```python
corelation = StudentPerformanceFactors.corr(numeric_only=True)
```

```python
sns.heatmap(corelation, xticklabels=corelation.columns, yticklabels=corelation.columns,annot=True) 
```

```python
#Heatmap Analysis: Attendance (0.58) and Study Hours (0.45) are the strongest factors positively influencing exam scores.Sleep Hours, Tutoring Sessions, and Physical Activity show minimal to no correlation with exam performance.
#Conclusion: Attendance and study hours are key to better exam results, while sleep, tutoring, and physical activity have little impact.
```

```python
#Distribution:-
```

```python
sns.pairplot(StudentPerformanceFactors)
```

```python
# Correlation between Hours Studied and Exam Score is positive, while other factors show weak relationships.
# Hours Studied and Previous Scores impact Exam Score, but Physical Activity and Tutoring Sessions have minimal effect.
```

```python
#Scatterplot:-
```

```python
sns.relplot(x='Exam_Score', y='Previous_Scores', hue='Gender', data=StudentPerformanceFactors)
```

```python
#Scatter Plot Analysis: There is no significant gender difference in exam scores, as both males and females perform similarly across different ranges.Previous Scores do not reliably predict current exam scores; performance varies more widely at lower previous score levels.
#Conclusion: Previous scores and gender do not strongly affect current exam performance.
```

```python
#Histogram:-
```

```python
sns.histplot(StudentPerformanceFactors['Gender'])
```

```python
#This is a bar plot showing the distribution of gender in the dataset, with more males than females.
#Conclusion: The dataset has a gender imbalance, with a higher count of male participants compared to female participants. This may need to be considered when interpreting the results to avoid bias.
```

```python
#Boxplot:-
```

```python
sns.catplot(x='Attendance', kind= 'box', data=StudentPerformanceFactors)
```

```python
#Median attendance: High (around 80%).
#Spread: Moderate.
#Outliers: None.
#Central Tendency: The median attendance is relatively high, indicating a majority of students are attending regularly.
#Conclusion: Students generally attend regularly with minimal variability.

```
